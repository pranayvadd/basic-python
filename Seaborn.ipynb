```python
sns.boxplot(data=tips,x='day',y='total_bill')
```

# Titanic 

```python
titanic = pd.read_csv('data/titanic.csv')

titanic.head()
```

```python
titanic.shape
```

```python
titanic.describe(include='all')
```

```python
sns.barplot(x='sex',y='age',data=titanic)
```

barplot tells you the mean of the column

male on an avg age is 31

female on an avg age is 27

```python
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
sns.barplot(x='sex',y='age',data=titanic)
plt.title('Avg age')

plt.subplot(1,2,2)
sns.barplot(x='sex',y='age',data=titanic,estimator=np.median)
plt.title('Median age')

plt.show()
```

above figure is to know avg age and median age of data

```python
sns.boxplot(x='sex',y='age',data=titanic)
```

Bivariate Analysis

Categorical colums 

Male has outliers that means , in titanic, old male people are there

```python
sns.barplot(x='sex',y='survived',data=titanic)
```

only 20% of the male survived

75% of the female survived

```python
sns.barplot(x='sex',y='survived',hue='class',data=titanic)
```
