```python
# check prime numbers 
%time

max_num = int(input('enter the number:'))

val = 2

count = 0

while val <= (int((max_num)**0.5) + 1) :
    if max_num % val == 0:
        count += 1
        break
    val += 1
else:
    print('out of loop')
    
if count == 0:
    print('prime')
else:
    print('not a prime')

print('STOP RUN')
```
