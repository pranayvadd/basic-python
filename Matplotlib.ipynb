```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
y = x ** 2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'b+')
plt.show()
```

```python
# Labelling

plt.plot(x,y,'r+',x,y,'k-')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot - x & y')

plt.xlim([0,10])
plt.ylim([0,100])

plt.show
```

```python
plt.plot([1,3,5,7,9], [1,8,3,5,1])
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Line plot - x & y')

plt.xlim([0,10])
plt.ylim([0,10])

plt.show()
```

```python
# Labelling the plot

heights = np.array([160, 180, 146, 162, 184, 180])
weights = np.array([50, 78, 45, 51, 80, 60])

plt.plot(heights, weights, 'r*')

plt.xlabel('Heights (cm)')
plt.ylabel('Weights (kg)')
plt.title('Scatter plot - Height & Weight')

plt.xlim([140,200])
plt.ylim([40,90])

plt.show
```

```python
# Multiple plots in a single figure

x = np.arange(1,21)
y_1 = np.log(x)
y_2 = np.exp(x)

#Inititaing a figure
plt.figure()

#Creating a plot with 1 row and 2 cols

plt.subplot(1,2,1)
plt.title('y= log(x)')
plt.plot(x, y_1)

plt.subplot(1,2,2)
plt.title('y= exp(x)')
plt.plot(x, y_2)

plt.show
```

Let us look at some common plots using matplotlib

1. Box PLOT
2. Scatter PLOT
3. Histograms

Statistics :

1. Univariate Analysis : 
    1. Measure of central tendency 
    2. Measure of spread
2. Bivariate Analysis  :
    1. Measure of relationship

Data VIZ :

1. Univariate Analysis : 
    1. Histograms
    2. Box plot
        1. Viz representation of the statistical description of one column at a time.
        2. Whisker : weather outliers exists in the data; length of Whisker  = 1.5 * IQR
        3. IQR (Inter Quartile Range) = 75%ile - 25%ile
        4. Upper limit = 75%ile + 1.5 * IQR
        5. Lower limit = 25%ile - 1.5 * IQR
        6. Outliers : xi > Upper limit or xi > Lower limit
        
        
    
2. Bivariate Analysis  : 
    1. Scatter plot
    2. Box plot
